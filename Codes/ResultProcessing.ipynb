```python
import pandas as pd
import numpy as np
import os
```

```python
files = os.listdir('Results/Cluster2')
full_result2 = pd.DataFrame()
for file in files:
    result = pd.read_csv('Results/Cluster2/' + file, index_col=0)
    full_result2 = pd.concat([result, full_result2])
full_result2.reset_index(inplace=True, drop=True)
```

```python
i = np.linspace(0, 27, 10).astype(int)
```

```python
full_result2.sort_values(by=['f1', 'bal_acc'], axis=0, ascending=False).iloc[i, :]
```

```python
full_result2.sort_values(by=['bal_acc', 'f1'], axis=0, ascending=False).iloc[i, :]
```
